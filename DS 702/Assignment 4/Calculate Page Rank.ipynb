### Without Taxation

```python
import numpy as np
from fractions import Fraction

M = [
    [(1/3), 0.5, 0],
    [(1/3), 0, 0.5],
    [(1/3), 0.5, 0.5],
]

M = np.asarray(M)

v = [(1/3), (1/3), (1/3)]
v = np.asarray(v)

M, v
```

```python
def toRatio(v):
    for i in v:
        print(str(Fraction(i).limit_denominator()), end='|')
toRatio(v)
```

```python
for j in range(10):
    print('Iteration {}'.format(j+1))
    print('-'*15)
    v = M@v
    print(v)
    toRatio(v)
    print('\n')
```

### With Taxation

```python
M = [
    [0, 0.5, 1, 0],
    [(1/3), 0, 0, 0.5],
    [(1/3), 0, 0, 0.5],
    [(1/3), 0.5, 0, 0]
]

# M = [
#     [(1/3), 0.5, 0],
#     [(1/3), 0, 0.5],
#     [(1/3), 0.5, 0.5],
# ]

# M = [
#     [0,    (1/4),  (1/4), (1/4), 0],
#     [(1/4), 0,     (1/4), (1/4), 0],
#     [(1/4), (1/4), 0,     (1/4), 0],
#     [(1/4), (1/4), (1/4), 0    , 0],
#     [(1/4), (1/4), (1/4), (1/4), 0]
# ]
M = np.asarray(M)

v = [(1/4), (1/4), (1/4), (1/4)]
# v = [(1/3), (1/3), (1/3)]
# v = [(1/5), (1/5), (1/5), (1/5), (1/5)]
v = np.asarray(v)

b = 0.8

e = (np.ones((M.shape[0],))/M.shape[0])*(1-b)
M= M*b
M, v, e
```

```python
for j in range(40):
    print('Iteration {}'.format(j+1))
    print('-'*15)
    v = M@v + e
    print(v)
    toRatio(v)
    print('\n')
```
